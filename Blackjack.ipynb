```python
import random

def deal_card():
    cards= [11,2,3,4,5,6,7,8,9,10,10,10,10]
    card = random.choice(cards)
    return card

def calculate_score(cards):
    if sum(cards) == 21 and len(cards) == 2:
        return 0
    if 11 in cards and sum(cards) > 21:
        cards.remove(11)
        cards.append(1)
    return sum(cards)

def compare(usr_sc, cpu_sc):
    if usr_sc == cpu_sc:
        return "Draw!"
    elif cpu_sc == 0:
        return "Computer wins, with a Blackjack!"
    elif usr_sc == 0:
        return "You win, with a Blackjack!"
    elif usr_sc > 21:
        return "You lose, you went over!"
    elif cpu_sc > 21:
        return "You win, computer went over!"
    elif usr_sc > cpu_sc:
        return "You win, you scored higher!"
    else:
        return "You lose."
    
def play_game():   
    user_card = []
    cpu_card = []
    game_over = False

    for _ in range(2):
        user_card.append(deal_card())
        cpu_card.append(deal_card())

    while not game_over:
        user_score = calculate_score(user_card)
        cpu_score = calculate_score(cpu_card)
        if user_score != 0:
            print(f"Your cards are {user_card}, your score is {user_score}")
        else:
            print(f"Your cards are {user_card}, your score is 21")

        print(f"The computer's first card is {cpu_card[0]}")

        if user_score == 0 or cpu_score == 0 or user_score > 21:
            game_over = True
        else:
            user_should_deal = input("Type 'y' to get another card, type 'n' to pass: ")
            if user_should_deal == "y":
                user_card.append(deal_card())
            else:
                game_over = True

    while cpu_score != 0 and cpu_score < 17:
        cpu_card.append(deal_card())
        cpu_score = calculate_score(cpu_card)
    print(f"The computer's cards were: {cpu_card}")
    print(f"Your final cards were: {user_card}")
    print(compare(user_score,cpu_score))

while input("Would you like to play a game of Blackjack? Type 'y' or 'n': ") == "y":
    play_game()
```
