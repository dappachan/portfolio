```python
import random
#with open("")
word_list = ["death","untitled","astronomy","python","kernel","database","breach","fortification","fanatic","wrist","grouper","harrowing","mother","wedding","matrimony","promise","tension","unethical",
             "unknown","fertility","frenetic","accomplice","unsaved","logout","password","management","spatula","coagulate","extraction","masculinity","femininity","needy","cloying",
             "hardworking","truthful","petition","weightlifting","action","behavior","disorder","spectrum","bias","urgency","dialect","territory","sedition","ambidextrous","district","apartheid",
             "guess","uncertainty","probability","dictionary","recollection","memory","notification","genocide","admission","repentence","biblical","opponent",
             "fearful","gestation","plantation","inheritance","wordful","wonderful","aspirational","therapeutic","hangman","cannibal","worthiness","wistful","fleeting","salutation",
             "realm", "oregano", "soundful", "sentence","wordage","verbage", "forceful","warning","lasagna","eggplant","barometer","murderous","beneficiary","domino","cauldron",
             "horizon","property","convenience","northern","ownership","gestapo","societal","exchange"]
counter = 0
answer = random.choice(word_list)
blank = []

for i in range(len(answer)):
    blank += "_"

print("This is a game of Hangman. You only get 20 attempts at guessing a letter before the man hangs. Guess the word at any time.")
while counter < 20:
    guess = input("Enter a letter(or your guess for the word): ").lower()
    blank_str = ""
    for letter in blank:
            blank_str = blank_str + letter + " "
    if guess in answer:
        if guess == answer:
            break
        for i in range(len(answer)):
            if guess == answer[i]:
                blank[i] = guess
        
        
        blank_str = ""
        for letter in blank:
            blank_str = blank_str + letter + " "
        print(blank_str)
    else:
        print("Uh oh!")
        print(blank_str)
    if blank == [*answer]:
        break
    counter += 1
    

if counter == 20:
    print("You lost! The word was " + answer)

else:
    print("Good job!")
```
